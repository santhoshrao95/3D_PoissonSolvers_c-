```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import csv
#from mpl_toolkits import mplot3d
import numpy as np
#import matplotlib.pyplot as plt
import math

with open('gs_33_5000_j=16.csv') as inf:
    phi = []
   # y1 = []
    for line in csv.reader(inf):
        tx =  line
        tx = [float(i) for i in tx]
        phi.append(tx)
#print(phi)


x = np.linspace(0, 1, 34)
z = np.linspace(0, 1, 34)

X, Z = np.meshgrid(x, z)
#phi = f(X, Z)
y = np.array(phi)
#print(phi)
ax = plt.axes(projection='3d')
ax.plot_surface(X,Z,y[:], rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
#ax.set_title('surface')


ax.set_xlim(0,1)
ax.set_ylim(0,1)
ax.set_xlabel('$x$')
ax.set_ylabel('$z$')
ax.set_zlabel('$phi$')
ax.set_title('GS_j=16_33*33*33')
plt.savefig('GS_j=16_33*33*33_3d.png')
ax.view_init(30,-70)
```
