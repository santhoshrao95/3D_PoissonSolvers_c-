```python
import csv
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt

#from mpl_toolkits import mplot3d
import numpy as np
#import matplotlib.pyplot as plt
import math

def f(x, z):
    
    return 500*np.exp(-50*((1-x)**2 + z**2)) + 100*x*(1-z)


x = np.linspace(0, 1, 33)
z = np.linspace(0, 1, 33)

X, Z = np.meshgrid(x, z)
phi1 = f(X, Z)
#print(phi)
ax = plt.axes(projection='3d')
ax.plot_surface(X,Z,phi1[:], rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('Exact_3d')


ax.set_xlim(0,1)
ax.set_ylim(0,1)
ax.set_xlabel('$x$')
ax.set_ylabel('$z$')
ax.set_zlabel('$phi$')
#ax.view_init(20,20)
plt.savefig('Exact_3D.png')
ax.view_init(30,-70)
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import math

def f(x, z):
    
    return 500*np.exp(-50*((1-x)**2 + z**2)) + 100*x*(1-z)


x = np.linspace(0, 1, 30)
z = np.linspace(0, 1, 30)

X, Z = np.meshgrid(x, z)
phi = f(X, Z)
fig = plt.figure(figsize=(6,5))
left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
ax = fig.add_axes([left, bottom, width, height])

cp = ax.contour(X, Z, phi)
ax.clabel(cp, inline=True,fontsize=10)
ax.set_title('Exact_countour')
plt.savefig('Exact_countour.png')
plt.show()
```
